```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set()
```

```python
dataset = pd.read_csv('Churn_Modelling.csv')
dataset.head()
```

```python
dataset = pd.get_dummies(dataset, columns=['Gender', 'Geography'], drop_first=True)
```

```python
dataset.head()
```

```python
from sklearn.preprocessing import StandardScaler
columns_scaled=['Gender_Male', 'Geography_Germany','Geography_Spain','CreditScore','Age','Tenure','Balance','EstimatedSalary']
scaler = StandardScaler()
dataset[columns_scaled] = scaler.fit_transform(dataset[columns_scaled])
```

```python
dataset
```

```python
dataset = dataset.loc[:,['RowNumber',
 'CustomerId',
 'Surname',
 'CreditScore',
 'Age',
 'Tenure',
 'Balance',
 'NumOfProducts',
 'HasCrCard',
 'IsActiveMember',
 'EstimatedSalary',
  'Gender_Male',
 'Geography_Germany',
 'Geography_Spain',
   'Exited'
]]
```

```python
#Customer ID and Surname would not be relevant as features
X_columns = dataset.columns.tolist()[3:14]
Y_columns = dataset.columns.tolist()[-1:]
print(X_columns)
print(Y_columns)
```

```python
X = dataset[X_columns].values
Y = dataset[Y_columns].values
```

```python
#Spilt the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
#Let us create the Neural Network
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
#Initialize ANN
classifier = Sequential()
```

```python
#Add input layer and hidden layer
classifier.add(Dense(6, activation = 'relu', input_shape = (X_train.shape[1], )))
classifier.add(Dropout(rate = 0.1))
```

```python
#Add second layer
classifier.add(Dense(6, activation = 'relu'))
classifier.add(Dropout(rate = 0.1))
```

```python
#Add output layer
classifier.add(Dense(1, activation = 'sigmoid'))
```

```python
#Let us take a look at our network
classifier.summary()
```

```python
#Optimize the weights
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
#Fitting the Neural Network
history = classifier.fit(X_train, y_train, batch_size = 32, epochs = 200, validation_split = 0.1, verbose = 2)
```

```python
y_pred = classifier.predict(X_test)
print(y_pred[:5])
```

```python
#Let us use confusion matrix with cutoff value as 0.5
y_pred = (y_pred > 0.5).astype(int)
print(y_pred[:5])
```

```python
#Making the Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
#Accuracy of our NN
print(((cm[0][0] + cm[1][1])* 100) / len(y_test), '% of data was classified correctly')
```
