```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sympy as sym
```

```python
def objective(x):
  return (x+3)**2
```

```python
def derivative(x):
  return 2*(x+3)
```

```python
def gradient_descent(alpha,start,max_iter,tolerance):
  x=start
  x_list=[]

  for i in range(max_iter):
    x_list.append(x)
    gradient=derivative(x)
    x=x - alpha*gradient

    # check
    if abs(gradient)<tolerance:
      break

  return x,x_list
```

```python
# hyperparameters
alpha=0.1
start=2
max_iter=30
tolerance=1e-6
```

```python
minima,x_list = gradient_descent(alpha,start,max_iter,tolerance)
```

```python
print('minima at: ',minima)
print('value: ',objective(minima))
```

PLOTS

```python
x=np.linspace(-10,5,100)
plt.plot(x,objective(x),label='obj function')

plt.plot(start,objective(start),'ro',label='initial point')
```

```python
x=np.linspace(-10,5,100)
plt.plot(x,objective(x),label='obj function')

plt.plot(start,objective(start),'ro',label='initial point')

x_arr=np.array(x_list)
plt.plot(x_arr,objective(x_arr),'.-',color='red', label='optimisation path')

plt.plot(minima,objective(minima),'go',label='minima pt')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Gradient Descent')
plt.legend()
plt.show()
```
